```python
train_data.drop(['id'],axis=1, inplace=True)
test_data.drop(['id'],axis=1, inplace=True)
```

```python
encoder = OrdinalEncoder()
train_data[["airline",	"flight",	"source_city",	"departure_time",	"stops",	"arrival_time",	"destination_city",	"class"]]=encoder.fit_transform(train_data[["airline",	"flight",	"source_city",	"departure_time",	"stops",	"arrival_time",	"destination_city",	"class"]])
```

```python
train_data.head()
```

```python
x = train_data.drop('price',axis=1).values
y = train_data['price'].values
```

```python
y
```

```python
linear_model = LinearRegression()
linear_model.fit(x,y)
```

```python
test_data[["airline",	"flight",	"source_city",	"departure_time",	"stops",	"arrival_time",	"destination_city",	"class"]]=encoder.fit_transform(test_data[["airline",	"flight",	"source_city",	"departure_time",	"stops",	"arrival_time",	"destination_city",	"class"]])
test_data = test_data.values
```

```python
predicted_test_data = linear_model.predict(test_data)
predicted_train_data = linear_model.predict(x)
```
